```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#importando dados 
#URL dos dados salvos no GITHUB
url = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
#Recupera os dados com pandas, informando que estão compactados
experimentos_df = pd.read_csv(url, compression='zip')
#Exibe uma prévia dos dados
experimentos_df.head(5)
```

##### DESAFIO 01 : INVESTIGAR POR QUE A CLASSE "TRATAMENTO" É TÃO DESBALANCEADA




```python
print('Verificando quais valores existem nesse campo')
print(experimentos_df['tratamento'].unique())
```

```python
print('Verificando o número de ocorrências para cada valor')
print(experimentos_df['tratamento'].value_counts())
```

```python
experimentos_controle = experimentos_df[experimentos_df['tratamento'] == 'com_controle']
print(experimentos_controle.head(5))
experimentos_controle['droga'].unique()
```

<p3>Confirmamos que o grupo de controle possui apenas "cacb2b860" como droga, o que pode ser um identificador informando que não é aplicada nenhuma droga</p3> 

```python
experimentos_droga = experimentos_df[experimentos_df['tratamento'] == 'com_droga']
print(experimentos_droga.head(5))
experimentos_droga['droga'].value_counts().head(15)
```
